```python
import tensorflow as tf
import numpy as np
import pandas as pd
from tensorflow import keras
from tensorflow.keras import layers
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Dropout
```

```python
#prepare the data
file_url="http://storage.googleapis.com/download.tensorflow.org/data/heart.csv"
```

```python
data=pd.read_csv(file_url)
```

```python
data.describe()
```

```python
data.columns
```

```python
from sklearn import preprocessing
le=preprocessing.LabelEncoder()
```

```python
data["thal"]=le.fit_transform(data["thal"])
data["thal"]
```

```python
x=data.iloc[:,:13].values
y=data.iloc[:,13:14].values
```

```python
x
```

```python
y
```

```python
data.dtypes
```

```python
data["thal"].nunique()
```

```python
#handling the features for NN model
from sklearn.preprocessing import StandardScaler,OneHotEncoder
sc=StandardScaler()
X=sc.fit_transform(x)
```

```python
#handling the label
ohe=OneHotEncoder()
Y=ohe.fit_transform(y).toarray()
```

```python
from sklearn.model_selection import train_test_split
```

```python
#set 20% for train and test for evaluation data
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.20,random_state=10)
```

```python
#set 25% for validation data
x_train,x_val,y_train,y_val=train_test_split(X_train,Y_train,test_size=0.25,random_state=10)
```

```python
model=Sequential()
model.add(Dropout(0.2))
model.add(Dense(32,input_shape=(13,),activation='relu',
                use_bias=True,
                kernel_initializer="glorot_uniform",
                bias_initializer="zeros",
                kernel_regularizer=None,
                bias_regularizer=None,
                activity_regularizer=None,
                kernel_constraint=None,
                bias_constraint=None))
model.add(Dense(2,activation='relu',kernel_initializer="glorot_uniform"))
```

```python
model.compile(optimizer="adam",loss="binary_crossentropy",metrics=["accuracy"])
```

```python
type(model)
```

```python
history=model.fit(X_train,Y_train,epochs=100,validation_data=([x_val,y_val]))
```
